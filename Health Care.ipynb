```python
enter = input("Enter the food groups: ")
for i in range(len(dataset)):
    if(dataset.FoodGroup.loc[i] == enter):
        activites = dataset.loc[i]
        labels = activites.index[1:]
        sizes = activites.iloc[1:]
        colors = ['orange', 'red', 'blue', 'yellowgreen', 'lightcoral', 'lightskyblue']
        plt.figure(figsize = (12, 8))
        plt.pie(sizes,labels=labels,autopct='%1.1f%%', shadow=True, startangle=140, colors = colors)
        plt.legend()
        plt.title('Nutritional Value Pie Chart')
        explode = (0.1, 0, 0, 0, 0, 0)
        plt.show()
        break
```
